```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
#Begin with the base code for sequential/linear and binary search 
#as described in chapter 1 of Grokking Algorithms (Bhargava 2016)

def simpleSearch(list, item):
    for i in list:
        if i == item:
            return i
    # Return none if item does not exist
    return None

def binarySearch(list, item):
  # Use low and high to set the boundaries in which the search will take place
  low = 0
  high = len(list) - 1

  # Use 'while' to keep the search going until you find the element
  while low <= high:
    # Check the element in the middle of low and high
    mid = (low + high) // 2
    guess = list[mid] #The middle element is our best guess
    # If the item is found, return it:
    if guess == item:
      return mid
    # If we guess too high, make the new high one step below the number we guessed
    if guess > item:
      high = mid - 1
    # If two low, make the new low one step above the number we guessed
    else:
      low = mid + 1

  # Return none if item does not exist
  return None
```

```python
#Set up random number arrays of 512, 1024, 2048, 4096, and 8192 using a uniform random number generator.
#Used a common, randomly generated seed of 572

np.random.seed(seed=572)

array512 = np.random.randint(9999, size=512)
array1024 = np.random.randint(9999, size=1024)
array2048 = np.random.randint(9999, size=2048)
array4096 = np.random.randint(9999, size=4096)
array8192 = np.random.randint(9999, size=8192)

```

```python
# Sort each of the random number arrays from smallest to largest.

start = time.clock()
array512 = np.sort(array512)
stop = time.clock()
sort512 = stop - start

start = time.clock()
array1024 = np.sort(array1024)
stop = time.clock()
sort1024 = stop - start

start = time.clock()
array2048 = np.sort(array2048)
stop = time.clock()
sort2048 = stop - start

start = time.clock()
array4096 = np.sort(array4096)
stop = time.clock()
sort4096 = stop - start

start = time.clock()
array8192 = np.sort(array8192)
stop = time.clock()
sort8192 = stop - start
```

```python
#Execute the base algorithm for each of the random number arrays, noting the execution time with each execution. 
#With each execution use the maximum value of the random number array as the number for which you are searching.

#Finsing all max values
search512 = max(array512)
search1024 = max(array1024)
search2048 = max(array2048)
search4096 = max(array4096)
search8192 = max(array8192)

# simple searches
start = time.clock()
simpleSearch(array512, search512)
stop = time.clock()
simple512time = stop - start

start = time.clock()
simpleSearch(array1024, search1024)
stop = time.clock()
simple1024time = stop - start

start = time.clock()
simpleSearch(array2048, search2048)
stop = time.clock()
simple2048time = stop - start

start = time.clock()
simpleSearch(array4096, search4096)
stop = time.clock()
simple4096time = stop - start

start = time.clock()
simpleSearch(array8192, search8192)
stop = time.clock()
simple8192time = stop - start

# Binary searches

start = time.clock()
binarySearch(array512, search512)
stop = time.clock()
binary512time = stop - start

start = time.clock()
binarySearch(array1024, search1024)
stop = time.clock()
binary1024time = stop - start

start = time.clock()
binarySearch(array2048, search2048)
stop = time.clock()
binary2048time = stop - start

start = time.clock()
binarySearch(array4096, search4096)
stop = time.clock()
binary4096time = stop - start

start = time.clock()
binarySearch(array8192, search8192)
stop = time.clock()
binary8192time = stop - start
```

```python
#Use Python (perhaps with the Pandas package) to prepare a six-column array showing 
#the length of the random number array in column 
#(1) and execution time in milliseconds in the remaining columns as follows: 
#(2) sort time, (3) linear search time for sorted array, (4) binary search time for sorted array, 
#(5) linear search plus sort times, and (6) binary search plus sort times.

# Creating the sort + search times
# Simple:
ss512 = sort512 + simple512time
ss1024 = sort1024 + simple1024time
ss2048 = sort2048 + simple2048time
ss4096 = sort4096 + simple4096time
ss8192 = sort8192 + simple8192time

#Binary:
bs512 = sort512 + binary512time
bs1024 = sort1024 + binary1024time
bs2048 = sort2048 + binary2048time
bs4096 = sort4096 + binary4096time
bs8192 = sort8192 + binary8192time

# Create array

searchSortPerformance = pd.DataFrame([["Length", "Sort Time", "Simple Search", "Binary Search",
                                 "Sort + Simple", "Sort + Binary"],
                                ["512", sort512*1000, simple512time*1000, binary512time*1000, ss512*1000, bs512*1000],
                                ["1024", sort1024*1000, simple1024time*1000, binary1024time*1000, ss1024*1000, bs1024*1000],
                                ["2048", sort2048*1000, simple2048time*1000, binary2048time*1000, ss2048*1000, bs2048*1000],
                                ["4096", sort4096*1000, simple4096time*1000, binary4096time*1000, ss4096*1000, bs4096*1000],
                                ["8192", sort8192*1000, simple8192time*1000, binary8192time*1000, ss8192*1000, bs8192*1000]])
searchSortPerformance
```

```python
#Use Python matplotlib or Seaborn to generate a plot with the size of the random 
#number array on the horizontal axis and with execution time in milliseconds on the vertical axis. 
#The plot should show execution time against array size for sequential and binary search algorithms alone.

plt.plot([512, 1024, 2048, 4096, 8192],
         [simple512time*1000, simple1024time*1000, simple2048time*1000, simple4096time*1000, simple8192time*1000],
         'm-',
         [512, 1024, 2048, 4096, 8192],
         [binary512time*1000, binary1024time*1000, binary2048time*1000, binary4096time*1000, binary8192time*1000],
         'b-')
plt.axis([0, 10000, 0, 1.6])
plt.xlabel('Array Size')
plt.ylabel('Execution Time in Milliseconds')
plt.title('Array Size vs Search Execution Time')
plt.legend(['Simple', 'Binary'], loc=2)
plt.show()
```

```python
#Use Python matplotlib or Seaborn to generate a measure of the size of the data set on the horizontal axis 
#and with execution time in milliseconds on the vertical axis. 
#The plot should show execution time against problem size for each form of the algorithm being tested.

plt.plot([512, 1024, 2048, 4096, 8192],
         [sort512*1000, sort1024*1000, sort2048*1000, sort4096*1000, sort8192*1000],
         'b-',
         [512, 1024, 2048, 4096, 8192],
         [simple512time*1000, simple1024time*1000, simple2048time*1000, simple4096time*1000, simple8192time*1000],
         'm-',
         [512, 1024, 2048, 4096, 8192],
         [binary512time*1000, binary1024time*1000, binary2048time*1000, binary4096time*1000, binary8192time*1000],
         'c-',
         [512, 1024, 2048, 4096, 8192],
         [ss512*1000, ss1024*1000, ss2048*1000, ss4096*1000, ss8192*1000],
         'm--',
         [512, 1024, 2048, 4096, 8192],
         [bs512*1000, bs1024*1000, bs2048*1000, bs4096*1000, bs8192*1000],
         'c--')
plt.axis([0, 10000, 0, 3])
plt.xlabel('Array Size')
plt.ylabel('Execution Time in Milliseconds')
plt.title('Array Size vs Search Execution Time with Sort Times')
plt.legend(['Numpy Sort', 'Simple', 'Binary', 'Sort + Simple', 'Sort + Binary'], loc=2)
plt.show()
```

### Executive Summary

In the above exercise, we implemented both simple and binary search on arrays of sizes 512, 1024, 2048, 4096, and 8192. Though searching may be a basic (and therefore overlooked) algorithm, it is an important and oft-used procedure when dealing with data. As data gets larger, it becomes increasingly important to be able to search efficiently.

As we can see in both the table and charts above, binary search far outperforms simple search. Interestingly,at array sizes above 4,000, binary search was even able to outperform the built-in Numpy sort method. In Big O notation, we can see that simple search performs at O(n),or linear time,  while binary search performs O(log(n)), or log time.

It is clear that binary search should be an understood and regularly implemented part of any data engineer's toolbox.
