```python
from numpy.random import seed
from numpy.random import randint
import numpy as np
import time
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Begin with the base code for recursion (factorial) as described in 
#chapter 3 of Grokking Algorithms (Bhargava 2016).

def fact(x):
    if x == 1:
        return 1
    else:
        return x * fact(x-1)
```
