```python
from numpy.random import seed
from numpy.random import randint
import numpy as np
import time
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Begin with the base code for recursion (factorial) as described in 
#chapter 3 of Grokking Algorithms (Bhargava 2016).

def fact(x):
    if x == 1:
        return 1
    else:
        return x * fact(x-1)
```

```python
#Calculate factorial using simple FOR loop and capture start/end time, 
#store the results in an array, or linked list, or dictionary

start = time.clock()
num = a0
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
stop = time.clock()
loop0time = stop - start

start = time.clock()
num = a1
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
stop = time.clock()
loop1time = stop - start

start = time.clock()
num = a2
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
stop = time.clock()
loop2time = stop - start

start = time.clock()
num = a3
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
stop = time.clock()
loop3time = stop - start

start = time.clock()
num = a4
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
stop = time.clock()
loop4time = stop - start

start = time.clock()
num = a5
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
stop = time.clock()
loop5time = stop - start

start = time.clock()
num = a6
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
stop = time.clock()
loop6time = stop - start

start = time.clock()
num = a7
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
stop = time.clock()
loop7time = stop - start

start = time.clock()
num = a8
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
stop = time.clock()
loop8time = stop - start

start = time.clock()
num = a9
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
stop = time.clock()
loop9time = stop - start

loopTimes = {'loop0' : round(loop0time,6), 'loop1' : round(loop1time,6), 
             'loop2' : round(loop2time,6),
         'loop3' : round(loop3time,6), 'loop4' : round(loop4time,6), 
             'loop5' : round(loop5time,6),
         'loop6' : round(loop6time,6), 'loop7' : round(loop7time,6), 
             'loop8' : round(loop8time,6),
         'loop9' : round(loop9time,6)}
loopTimes
```

```python
#Organize the results of the study into a table showing the number,
#it's factorial, the time to calculate the factorial using recursion, 
#the time to calculate factorial using FOR loop, and the difference 
#between times i.e. factorial vs. loop.
from decimal import Decimal

myResults = pd.DataFrame(
                        [[a0, "{:.2e}".format(Decimal(fact(a0))), a0time, loop0time],
                         [a1, "{:.2e}".format(Decimal(fact(a1))), a1time, loop1time],
                         [a2, "{:.2e}".format(Decimal(fact(a2))), a2time, loop2time],
                         [a3, "{:.2e}".format(Decimal(fact(a3))), a3time, loop3time],
                         [a4, "{:.2e}".format(Decimal(fact(a4))), a4time, loop4time],
                         [a5, "{:.2e}".format(Decimal(fact(a5))), a5time, loop5time],
                         [a6, "{:.2e}".format(Decimal(fact(a6))), a6time, loop6time],
                         [a7, "{:.2e}".format(Decimal(fact(a7))), a7time, loop7time],
                         [a8, "{:.2e}".format(Decimal(fact(a8))), a8time, loop8time],
                         [a9, "{:.2e}".format(Decimal(fact(a9))), a9time, loop9time]],
columns = ['Number', 'Factorial', 'Calculation Time (Recursion)',
                          'Calculation Time (For-Loop)'])
myResults['RunTime Difference (Loop - Recursion)'] = myResults['Calculation Time (For-Loop)'] - myResults['Calculation Time (Recursion)']
myResults
```
