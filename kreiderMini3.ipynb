```python
from numpy.random import seed
from numpy.random import randint
import numpy as np
import time
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Begin with the base code for recursion (factorial) as described in 
#chapter 3 of Grokking Algorithms (Bhargava 2016).

def fact(x):
    if x == 1:
        return 1
    else:
        return x * fact(x-1)
```

```python
#Set up an array of ten random numbers between 100 and 500 (just one array with 10 numbers).

seed(7769)
myArray = list(randint(100, 500, 10))
myArray.sort()


print(myArray)

```

```python
#Organize the results of the study into a table showing the number,
#it's factorial, the time to calculate the factorial using recursion, 
#the time to calculate factorial using FOR loop, and the difference 
#between times i.e. factorial vs. loop.
from decimal import Decimal

myResults = pd.DataFrame(
                        [[a0, "{:.2e}".format(Decimal(a0fact)), a0time, loop0time],
                         [a1, "{:.2e}".format(Decimal(a1fact)), a1time, loop1time],
                         [a2, "{:.2e}".format(Decimal(a2fact)), a2time, loop2time],
                         [a3, "{:.2e}".format(Decimal(a3fact)), a3time, loop3time],
                         [a4, "{:.2e}".format(Decimal(a4fact)), a4time, loop4time],
                         [a5, "{:.2e}".format(Decimal(a5fact)), a5time, loop5time],
                         [a6, "{:.2e}".format(Decimal(a6fact)), a6time, loop6time],
                         [a7, "{:.2e}".format(Decimal(a7fact)), a7time, loop7time],
                         [a8, "{:.2e}".format(Decimal(a8fact)), a8time, loop8time],
                         [a9, "{:.2e}".format(Decimal(a9fact)), a9time, loop9time]],
columns = ['Number', 'Factorial', 'Calculation Time (Recursion)',
                          'Calculation Time (For-Loop)'])
myResults['RunTime Difference (Loop - Recursion)'] = myResults['Calculation Time (For-Loop)'] - myResults['Calculation Time (Recursion)']
myResults
```

### Executive Summary

It is often the case that runtime efficiency is of paramount importance when deciding between two methods in data science; however, it is certainly not the only consideration. Ease of creation, code evaluation, and reproducibility all factor in to the decision on which method works best in each situation.

As we can see in the above chart, our recursive and for-loop algorithms perform similarly at the low and high ends of the spectrum of numbers, but differ greatly in the center of the range. The recursive algorithm loses efficiency and sees a spike in runtimes, while the for-loop sees a dip. In Big O notation, they both generally perform at O(n), but not in a completely linear way. 

In the table above, we see that the for-loop outperforms the recursive in all but four of our samples. Due to this, I would recommend implementing the for-loop version for calculating factorials, as the loop is simple enough to not add any undue complexity to understanding the overall code.
